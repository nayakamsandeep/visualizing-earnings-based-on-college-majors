```python
import pandas as pd
import matplotlib

recent_grads = pd.read_csv('recent-grads.csv')

print(recent_grads.iloc[0])
print(recent_grads.head())
print(recent_grads.tail())
recent_grads.describe()
```

```python
raw_data_count = len(recent_grads)
recent_grads = recent_grads.dropna(axis = 0)
cleaned_data_count = len(recent_grads)

print(raw_data_count)
print(cleaned_data_count)
```

```python
import matplotlib.pyplot as plt

fig, (ax1, ax2, ax3, ax4, ax5, ax6) = plt.subplots(6, figsize=(20,60))


ax1.scatter(recent_grads['Sample_size'], recent_grads['Median'])
ax1.set_ylim(18000,80000)
ax1.set_xlim(-500,3000)
ax1.set_title('Sample_size Vs Median')
ax1.set_xlabel('Sample_size')
ax1.set_ylabel('Median')

ax2.scatter(recent_grads['Sample_size'], recent_grads['Unemployment_rate'])
ax2.set_xlim(-500,1000)
ax2.set_ylim(0, 0.20)
ax2.set_title('Sample_size Vs Unemployment_rate')
ax2.set_xlabel('Sample_size')
ax2.set_ylabel('Unemployment_rate')

ax3.scatter(recent_grads['Full_time'], recent_grads['Median'])
ax3.set_xlim(-25000,160000)
ax3.set_ylim(18000, 80000)
ax3.set_title('Full_time Vs Median')
ax3.set_xlabel('Full_time')
ax3.set_ylabel('Median')

ax4.scatter(recent_grads['ShareWomen'], recent_grads['Unemployment_rate'])
ax4.set_title('ShareWomen Vs Unemployment_rate')
ax4.set_xlabel('ShareWomen')
ax4.set_ylabel('Unemployment_rate')

ax5.scatter(recent_grads['Men'], recent_grads['Median'])
ax5.set_xlim(-25000,160000)
ax5.set_ylim(18000, 80000)
ax5.set_title('Men Vs Median')
ax5.set_xlabel('Men')
ax5.set_ylabel('Median')

ax6.scatter(recent_grads['Women'], recent_grads['Median'])
ax6.set_xlim(-25000,180000)
ax6.set_ylim(18000, 80000)
ax6.set_title('Wome Vs Median')
ax6.set_xlabel('Women')
ax6.set_ylabel('Median')

plt.show()
```

```python
fig, (ax1, ax2, ax3, ax4, ax5, ax6, ax7, ax8) = plt.subplots(8,figsize = (10, 50))


column_list = ['Sample_size', 'Median', 'Employed', 'Full_time', 'ShareWomen', 'Unemployment_rate', 'Men', 'Women']
axis_list = [ax1, ax2, ax3, ax4, ax5, ax6, ax7, ax8]
for row in range(0,8):
    axis_list[row].hist(recent_grads[column_list[row]])
    axis_list[row].set_xlabel(column_list[row])
    axis_list[row].set_ylabel('Frequency')
```

```python
from pandas.tools.plotting import scatter_matrix

scatter_matrix(recent_grads[['Sample_size', 'Median']])
plt.show()
```
